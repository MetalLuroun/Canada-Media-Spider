```python
df = globalNewsDropOld(df)
print(len(df))
```

```python
df['emotion'] = 1
df ['subjectivity'] = 1
df.head(3)
df['content'] = df['content'].astype('str')
```

```python
i = 1
for index, row in df.iterrows():
#     print(row['content'])
    if not row['content'] == "Not Text Type":
        print("processing the " + str(i) + "article")
        i += 1 
        df.loc[index , 'emotion']   = TextBlob(row['content']).sentiment.polarity
        df.loc[index , 'subjectivity']   = TextBlob(row['content']).sentiment.subjectivity
#         print(TextBlob(row['content']).sentiment.polarity)
        continue
    df.loc[index , 'emotion']  = 0
```
