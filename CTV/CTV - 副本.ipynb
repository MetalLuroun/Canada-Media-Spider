```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
searchword = input("Enter your keyword you want to search on CTV : ")
```

```python
def getHTML(keyword , i):  
    url = "https://www.ctvnews.ca/search-results/search-ctv-news-7.137?page=" + str(i) + "&sortOrder=date&q="+keyword+"&fdate=&ftype=&fpage="
    payload={}
    headers = {}
    proxy = {"http":"http://127.0.0.1:7890"}
    try:
        response = requests.request("GET", url, headers=headers ,  proxies = proxy)
        return response
    except:
        print('Connect Failed')
        pass
    
```

```python
class Article:
    times = 0
    def __init__(self):
        self.title = ""
        self.description = ""
        self.time = ""
        self.url = ""
        self.content = ""
        Article.times = Article.times + 1
    def printOne(self):
        print("THIS IS " + str(Article.times) + " ARTICLE")
        print("TITLE : "+self.title)
        print("DESCRIPTION : "+self.description)
        print("TIME : "+self.time)
        print("URL : " +self.url)
#         print("CONTENT : " +self.content)
        print("-----------------------------------------------------")
    def getContent(self):
        print("getting content from "+ self.url + "......")
        proxy = {"http":"http://127.0.0.1:7890"}
        contentResponse = requests.request("GET" , self.url , proxies = proxy )
        soup = BeautifulSoup(contentResponse.text , "html.parser")
        try:
            for i in soup.find(class_ = "articleBody").find_all("p"):
                self.content += i.get_text()
        except:
            pass
    def to_csv(self):
        try:
            with open(searchword+'.csv', 'w', newline='' , encoding = "utf-8") as csvfile:
                CSVwriter = csv.writer(csvfile, delimiter=',', quoting=csv.QUOTE_MINIMAL)
                CSVwriter.writerow([self.title] + [self.description] + [self.time] + [self.url] + [self.content])
        except:
            print('fail to write to CSV')
            pass
```

```python
def getInfo(html): 
    soup = BeautifulSoup(html , "html.parser")
    articles =  soup.find_all("li" , "searchHit")
#     titles = articles.find_all('h3')
#     descriptions = article.find_all(class_ = "text")
#     times = articles.find_all(class_ = "meta")
#     urls = articles.find_all(class_ = "article-card__link")

    for i in range(0,10):
        article = Article()
        article.title = articles[i].h3.get_text()
        article.description = articles[i].find(class_="text").get_text()
        article.time = articles[i].find(class_ = "meta").get_text()
        article.url = articles[i].a['href']
        article.getContent()
        article.printOne()
        try:
            with open(searchword+'.csv', 'a', newline='' , encoding = "utf-8") as csvfile:
                cbc_China = csv.writer(csvfile, delimiter=',', quoting=csv.QUOTE_MINIMAL)
                cbc_China.writerow([article.title]  + [article.description] + [article.time] + [article.url] + [article.content])
        except:
            pass
```
